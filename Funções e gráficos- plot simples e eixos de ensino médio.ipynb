### Importando bibliotecas

```python
from math import *    # Importando as bibliotecas de operações matemáticas
import numpy as np    # Importando as bibliotecas de numpy
import matplotlib.pyplot as plt  # Importando as bibliotecas para geras gráficos
```

## Definindo função para uma reta:
>### Aqui faremos uma reta partindo de uma função com escolhidos coeficientes angular e linear, dependente da variável x: $$f(x,a,b) = ax + b$$

```python
def reta(x,a,b): # a coef angular, b é coef. linear

    f = a*x + b  # a identação indica o que é definido dentro da função, f é só uma variável de retorno(não existe fora da função)
    return f     # o retorno indica o que a função devolve quando é chamada
```

## Definindo função para uma parábola: 
>### Semelhante ao caso anterior, mas aqui são dados três valores: $$f(x,a,b,c) = ax^2 +bx + c$$

```python
def parab(x,a,b,c):     #aqui a função possui 4 variáveis, como mostrado a seguir as variáveis podem ser multidimensionais (arrays)
    
    return a*x*x + b*x + c   #em casos simples como esse é possível retornar diretamente a operação com as variáveis
```

# Plotando com o auxílio de funções

```python
plt.figure()                 #essa linha sempre deve começar no plot do gráfico, ela indica que estamos abrindo a figura

plt.plot([-3,5],[-3,7])      #o comando plot liga os pontos de duas listas ou arrays, o primeiro argumento é das coordenadas
                             #em x e o segundo argumento é das coordenadas em y
    
plt.show()                   #mostra a figura
plt.close()                  #fecha a figura (mais útil em loops)
```

```python
plt.figure()
plt.plot([-3,5,-2],[-3,7,5])    # O comando plot liga dois pontos consecutivos dado pelas entradas
plt.show()
plt.close()
```

```python
x1 = np.linspace(-5,10,100)        # comando que cria um conjunto de pontos igualmente espaçados dentro de um intervalo
                                   # np.linspace(começo do intervalo, fim do intervalo, total de pontos)
    
parabola1 = parab(x1, 1, -6, 4)    # faz com que a variável parabola1 guarde um conjunto de 100 pontos que são resposta da 
                                   # função parábola escolhida.

plt.figure()
plt.plot(x1,parabola1, color = 'black')   # No lugar das listas, colocamos x1 para coordenadas em x e parabola1 para
                                          # coordenadas em y
plt.show()
```

# Plot de circunferências e elipses:
## Para fazer esses plots, usamos auxílio de equações paramétricas:
>
>## Circunferência:
>># $$\begin{cases}x(t) = R cos(t) + x_c \\ y(t) = R sen(t) +y_c   \end{cases}$$
>
>## Elipse:
>># $$\begin{cases}x(t) = a cos(t) + x_c \\ y(t) = b sen(t) + y_c   \end{cases}$$


```python
# Definindo a circunferência
def circ(t,R):
    
    x = R* np.cos(t)
    y = R * np.sin(t)
    
    return x, y   #Aqui são dados dois retornos, que indicam os pontos resultantes para as coordenadas x e y
```

```python
t = np.linspace(0, 2 *np.pi, 100)    #espaço de pontos que vai de 0 a 2pi, para uma circunferência completa
circ(t,5)
```

```python
plt.figure(figsize = (5,5))         # para deixar o círculo em escala, deixamos a figura quadrada

P = circ(t,5)

plt.plot(P[0], P[1])

plt.show()
```

```python
# Definindo a elipse
def elip(t,a,b):
    
    x = a* np.cos(t)
    y = b * np.sin(t)
    
    return x, y
```

```python
plt.figure(figsize = (20,4))

Q = elip(t,10,2)

plt.plot(Q[0], Q[1])

plt.show()
```
