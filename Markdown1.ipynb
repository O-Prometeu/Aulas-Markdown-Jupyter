## Deslocamento por tópicos:

### Quando temos um título correlacionado com um assunto é possível deslocar levemente textos ou imagens com o auxílio de um ou mais '>', como podemos ver a seguir:

## Título A
> ### Assunto B (relacionado com A)
>> #### Assunto C (Relacionado com B)
>> #### Assunto D(Relacionado com B no mesmo grupo de C)
>
>> #### Assunto E(Relacionado com B mas não no mesmo grupo de C e D)

## Título A
> ### Assunto B (relacionado com A)
>> #### Assunto C (Relacionado com B)
>> #### Assunto D(Relacionado com B no mesmo grupo de C)
>
>> #### Assunto E(Relacionado com B mas não no mesmo grupo de C e D)
